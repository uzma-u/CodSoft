```python
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv('advertising.csv')
LABELS = ["TV", "Sales"]

count_classes = pd.value_counts(data['TV'], sort = True)
count_classes.plot(kind = 'bar', rot=0)
plt.title("Transaction")
plt.xticks(range(2), LABELS)
plt.xlabel("TV")
plt.ylabel("Sales");
```

```python
data.TV.value_counts()
```

```python
TV = data[data['TV']==1]
Newspaper = data[data['Newspaper']==0]

print(f"Shape of TV adds: {TV.shape}")
print(f"Shape of Newspaper adds: {Newspaper.shape}")
```

```python
plt.figure(figsize=(14, 12))

plt.subplot(2, 2, 1)
data[data.TV == 1].Sales.hist(
  bins=35, color='red', alpha=0.6, label="Sales with Tv adds"
)
plt.legend()

plt.subplot(2, 2, 2)
data[data.Newspaper == 1].Sales.hist(
  bins=35, color='teal', alpha=0.6, label="Sales with Newspaper adds"
)
plt.legend()
```
